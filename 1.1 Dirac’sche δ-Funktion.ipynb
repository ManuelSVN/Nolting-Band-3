```python
import numpy as np
import matplotlib.pyplot as plt

a = 10
n = 1
x = np.linspace(0, 30, 100)
y = n / (np.pi * n**2 + (x - a)**2)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Lorentz Curve')
plt.title('Plot of Lorentz Curve')
plt.grid(True)
plt.show()
```

```python
from scipy.integrate import quad
def lorentz_curve(x, a, n):
    return n / (np.pi * n**2 + (x - a)**2)

a = 10
n = 0.0001
x = np.linspace(0, 30, 1000)
y = n / (np.pi * n**2 + (x - a)**2)
plt.plot(x, y)
y = lorentz_curve(x, a, n)

# Upper bound for the range of integration
b = 10.1

# Calculate the area under the curve from 0 to b using quad function
area_under_curve, _ = quad(lorentz_curve, 9.9, b, args=(a, n))

print("Area under the curve from 0 to", b, ":", area_under_curve)
```

### Funktion von Wiki
$$
\delta(x) = \frac{1}{{\sqrt{\pi \cdot a}}} \cdot e^{-\frac{x^2}{a}}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_pdf(x, a):
    return (1 / np.sqrt(np.pi * a)) * np.exp(-x**2 / a)

# Generating x values from -5 to 5 with step 0.1
x = np.arange(-5, 5, 0.1)

# Setting the value of 'a'
a = 1.0

# Calculating the corresponding y values using the Gaussian PDF formula
y = gaussian_pdf(x, a)

# Plotting the Gaussian curve
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Gaussian Probability Density Function')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy import integrate

def gaussian_pdf(x, a):
    return (1 / np.sqrt(np.pi * a)) * np.exp(-x**2 / a)

# Constants
a = 1.0
x1 = -4.0
x2 = 4.0

# Perform numerical integration using scipy
result, _ = integrate.quad(gaussian_pdf, x1, x2, args=(a))

print("Integral of Gaussian PDF from x = {} to x = {}: {:.6f}".format(x1, x2, result))
```
