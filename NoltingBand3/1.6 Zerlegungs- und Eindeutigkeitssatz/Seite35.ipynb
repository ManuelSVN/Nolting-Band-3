```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values
x = np.linspace(0, 2 * np.pi, 100)  # This generates 100 points from 0 to 2*pi

# Calculate the corresponding y values using the cosine function
y = np.cos(x)

# Create the plot
plt.figure(figsize=(8, 4))  # Optional: Adjust the figure size
plt.plot(x, y, label='cos(x)')

# Customize the plot
plt.title('Cosine Function')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
