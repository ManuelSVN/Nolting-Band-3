$$ \vec{F} = (x^2 y, x + y, y z)$$


```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the grid of points in the x, y, and z directions
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
z = np.linspace(-2, 2, 10)
X, Y, Z = np.meshgrid(x, y, z)

# Calculate the vector field at each point
U = X**2 * Y
V = X + Y
W = Y * Z

# Create a 3D quiver plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=0.1, normalize=True)

# Set plot labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Vector Field: $\\vec{F} = (x^2y, x + y, yz)$')

# Show the plot
plt.show()
```

```python
from sympy import symbols, diff
from sympy.vector import CoordSys3D

# Define the symbols
x, y, z = symbols('x y z')

# Define the coordinate system
N = CoordSys3D('N')

# Define the vector field
F = x**2 * y * N.i + (x + y) * N.j + y * z * N.k

# Calculate the divergence of the vector field
div_F = diff(F.dot(N.i), x) + diff(F.dot(N.j), y) + diff(F.dot(N.k), z)
div_F.simplify()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Creating a meshgrid for the x, y coordinates
x, y = np.linspace(-2, 2, 10), np.linspace(-2, 2, 10)
x, y = np.meshgrid(x, y)
z = 2*x*y + y + 1  # Divergence of the vector field

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plotting the surface
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Divergence')
ax.set_title('Divergence of the Vector Field')

# Adding a color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

plt.show()
```

```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Creating a meshgrid for plotting
x_range = np.linspace(-2, 2, 10)
y_range = np.linspace(-2, 2, 10)
z_range = np.linspace(-2, 2, 10)
x, y, z = np.meshgrid(x_range, y_range, z_range)

# Defining the components of the vector field for numpy arrays
U = x**2 * y
V = x + y
W = y * z

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plotting the vectors
ax.quiver(x, y, z, U, V, W, length=0.1, normalize=True)

# Labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Vector Field')

plt.show()
```
