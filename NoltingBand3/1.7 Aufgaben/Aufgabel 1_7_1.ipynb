```python
import matplotlib.pyplot as plt
import numpy as np

# Function as given in the image
def f_n(x, a, n):
    return (1 / np.sqrt(np.pi * n)) * np.exp(-(x - a)**2 / n)

# Parameters for the plot
a = 0  # Center of the peak
n_values = [0.1, 0.01, 0.001, 0.0001]  # Different values of n to show the limit
def new_func():
    x = np.linspace(-2, 2, 400)
    return x

x = new_func()  # Range of x values

# Plot the function for different values of n
plt.figure(figsize=(10, 6))

for n in n_values:
    plt.plot(x, f_n(x, a, n), label=f'n={n}')

plt.title('Function plots for different values of n approaching 0')
plt.xlabel('x')
plt.ylabel('f_n(x)')
plt.legend()
plt.grid(True)
plt.show()
```

Zu $\alpha < a <\beta$ : 
$$ \int_{-\infty}^{\infty} e^{-y^2} \, dy $$

```python
from sympy import symbols, exp, integrate, oo

# Defining the variable
y = symbols('y')

# Defining the integral
integral = integrate(exp(-y**2), (y, -oo, oo))

integral.evalf()
```

### Plot of the function:  
$$ e^{-y^2} $$

```python
import matplotlib.pyplot as plt
import numpy as np

# Creating a range of values for y
y = np.linspace(-2, 2, 400)

# Calculating the function values
f_y = np.exp(-y**2)

# Plotting the function
plt.figure(figsize=(8, 6))
plt.plot(y, f_y, label=r'$e^{-y^2}$')
plt.xlabel('y')
plt.ylabel('f(y)')
plt.title('Plot of the function $e^{-y^2}$')
plt.legend()
plt.grid(True)
plt.show()
```
