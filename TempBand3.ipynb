## Plot Beispiel Papula Formelsammlung (div):
$$
\vec{F}=\left(\begin{array}{c}
x^2 y \\
x+y \\
y z
\end{array}\right)
$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the grid of points in the x, y, and z directions
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
z = np.linspace(-2, 2, 10)
X, Y, Z = np.meshgrid(x, y, z)

# Calculate the vector field at each point
U = X
V = Y
W = Z

# Create a 3D quiver plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=0.1, normalize=True)

# Set plot labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Vector Field: $\\vec{F} = (x^2y, x + y, yz)$')

# Show the plot
plt.show()
```

```python
import sympy as sp

# Define the variables
x, y, z = sp.symbols('x y z')

# Define the vector field
F = sp.Matrix([x*y*z, x+y, z**2])


# Compute the curl (rotation) of the vector field
curl_F = sp.Matrix([F[2].diff(y) - F[1].diff(z), F[0].diff(z) - F[2].diff(x), F[1].diff(x) - F[0].diff(y)])

# Display the result as a vector
print(curl_F)
```

```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the grid of points in the x, y, and z directions
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
z = np.linspace(-2, 2, 10)
X, Y, Z = np.meshgrid(x, y, z)

# Calculate the vector field at each point
U = 0
V = X*Y
W = -X*Z + 1

# Create a 3D quiver plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=0.1, normalize=True)

# Set plot labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Vector Field: $\\vec{F} = (x^2y, x + y, yz)$')

# Show the plot
plt.show()
```
